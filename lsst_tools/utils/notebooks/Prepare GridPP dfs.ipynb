```python
import lsst_tools as lsstt
```

```python
opsimdbpath = os.environ.get('OPSIMDBPATH')

print(opsimdbpath)
engine = create_engine('sqlite:///' + opsimdbpath)
```

```python
opsimdfWFD = pd.read_sql_query('SELECT * FROM Summary WHERE fieldID >= 1000 AND fieldID < 1005', engine)
opsimdfDDF = pd.read_sql_query('SELECT * FROM Summary WHERE fieldID = 2786', engine)
```

```python
opsimdfWFD = lsstt.utils.find_MW_extinction(opsimdfWFD)
opsimdfDDF = lsstt.utils.find_MW_extinction(opsimdfDDF)
```

```python
opsimdfWFD[['obsHistID', 'filter', 'expMJD',
        'fieldID', 'fieldRA', 'fieldDec','fiveSigmaDepth', "EBV_MW"]]
```

```python
opsimdfDDF[['obsHistID', 'filter', 'expMJD',
        'fieldID', 'fieldRA', 'fieldDec','fiveSigmaDepth', "EBV_MW"]]
```

```python
filter_names = ["LSST_u","LSST_g","LSST_r","LSST_i","LSST_z","LSST_y"]

zp_dict = {}
for i in filter_names:
    zp_dict[i] = pcc.kcorr.calc_AB_zp(i)
#     zp_dict[pcc.utils.b(i)] = pcc.kcorr.calc_AB_zp(i)    
```

```python
# opsimdfDDF["long_filter"] = opsimdfDDF["filter"].apply(lsstt.sims.string_format_for_mapping, args=("LSST_",))
# opsimdfWFD["long_filter"] = opsimdfWFD["filter"].apply(lsstt.sims.string_format_for_mapping, args=("LSST_",))

opsimdfDDF["long_filter"] = opsimdfDDF["filter"].map(lambda x : "LSST_"+x)
opsimdfWFD["long_filter"] = opsimdfWFD["filter"].map(lambda x : "LSST_"+x)
```

```python
opsimdfDDF["zp"] = opsimdfDDF["long_filter"].map(zp_dict)
opsimdfWFD["zp"] = opsimdfWFD["long_filter"].map(zp_dict)
```

```python
opsimdfDDF[['obsHistID', 'long_filter', 'expMJD',
        'fieldID', 'fieldRA', 'fieldDec','fiveSigmaDepth', "EBV_MW", "zp"]]
```
